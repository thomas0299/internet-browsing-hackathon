## Making synthetic data

This is how we want our data to look like:

|User ID|Week|Google Searches|Emissions from Google searches|Emails sent|Emissions from emails|Time spent browsing|Emissions from browsing|
|---|---|---|---|---|---|---|---|
|integer|integer|number of|kg Co2 equiv.|number of|kg Co2 equiv.|hours|kg Co2 equiv.|

```python
import pandas as pd
import numpy as np

week = pd.DataFrame(np.random.randint(0,52,size=(100000, 1)), columns=['week'])
```

```python
google_search=pd.DataFrame(np.random.randint(0,800,size=(100000, 1)), columns=['google_search'])
```

```python
emails_sent=pd.DataFrame(np.random.randint(0,300,size=(100000, 1)), columns=['emails_sent'])
```

```python
browsing_time=pd.DataFrame(np.random.randint(0,70,size=(100000, 1)), columns=['browsing_time'])
```

```python
user_id=pd.DataFrame(np.random.randint(0,200,size=(100000, 1)), columns=['user_id'])
```

```python
fake_df=pd.concat([week, google_search,emails_sent, browsing_time, user_id], axis=1)

fake_df
```

```python
fake_df['google_emissions']=fake_df['google_search']*0.005
fake_df['emails_emissions']=fake_df['emails_sent']*0.004
fake_df['browsing_emissions']=fake_df['browsing_time']*0.036
```

```python
user_df=fake_df.groupby(['user_id', 'week']).sum()

user_df.reset_index(inplace=True)

user_df
```
